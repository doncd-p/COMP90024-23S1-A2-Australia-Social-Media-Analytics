## Income

```python
# Set filepath
fp = "../data/raw/sudoData/abs_2021census_g33_aust_lga-239718880597863891.csv"
# read income
income = pd.read_csv(fp)
income.columns
```

```python
len(income)
```

```python
cols = [" lga_code_2021", " negative_nil_income_tot", " hi_1_149_tot", " hi_150_299_tot", " hi_300_399_tot", " hi_400_499_tot",
        " hi_500_649_tot", " hi_650_799_tot", " hi_800_999_tot", " hi_1000_1249_tot", " hi_1250_1499_tot", " hi_1500_1749_tot", 
        " hi_1750_1999_tot", " hi_2000_2499_tot", " hi_2500_2999_tot", " hi_3000_3499_tot", " hi_3500_3999_tot", " hi_4000_more_tot",
        " tot_tot"]
income = income[cols]
```

```python
# remove the space in the column names 
income.columns = income.columns.str.replace(' ', '')
income
```

```python
from statistics import mean 
# average for 'negative_nil_income_tot' is set to be 0.
# average for 'hi_4000_more_tot' is set to be 4000.
# average for other range of income is set to be the mean of the range.
income['avg_weekly_income'] = (income['hi_1_149_tot'] * mean([1, 149]) 
                            + income['hi_150_299_tot'] * mean([150,299]) 
                            + income['hi_300_399_tot'] * mean([300,399])
                            + income['hi_400_499_tot'] * mean([400,499])
                            + income['hi_500_649_tot'] * mean([500,649])
                            + income['hi_650_799_tot'] * mean([650,799])
                            + income['hi_800_999_tot'] * mean([800,999])
                            + income['hi_1000_1249_tot'] * mean([1000,1249])
                            + income['hi_1250_1499_tot'] * mean([1250,1499])
                            + income['hi_1500_1749_tot'] * mean([1500,1749])
                            + income['hi_1750_1999_tot'] * mean([1750,1999])
                            + income['hi_2000_2499_tot'] * mean([2000,2499])
                            + income['hi_2500_2999_tot'] * mean([2500,2999])
                            + income['hi_3000_3499_tot'] * mean([3000,3499])
                            + income['hi_3500_3999_tot'] * mean([3500,3999])
                            + income['hi_4000_more_tot'] * 4000) / income['tot_tot']
income
```

```python
income = income[['lga_code_2021', 'avg_weekly_income']]
income
```

```python
# save to processed dataset
income.to_csv('../data/curated/sudoData/income.csv', index=False)
```

## Education Level

```python
# Set filepath
fp = "../data/raw/sudoData/lga_p16a_highest_yr_sch_by_age_by_sex_census_2016-1816960792523127682.csv"
# read education level
eduLevel = pd.read_csv(fp)
eduLevel.columns
```

```python
edu_cols = [" lga_code_2016", " f_dngts_tot", " p_y10e_tot", " p_y11e_tot", " p_y12e_tot", " p_y9e_tot", " m_dngts_tot",' m_y8b_tot',' f_y8b_tot']
edu = eduLevel[edu_cols]
# remove the space in the column names 
edu.columns = edu.columns.str.replace(' ', '')
# sum and remove female/male year 8 or below
edu['p_y8b_tot'] = edu['f_y8b_tot'] + edu['m_y8b_tot']
edu = edu.drop(columns=['f_y8b_tot','m_y8b_tot'])
# sum and remove female/male did not go to school columns
edu['p_dngts_tot'] = edu['f_dngts_tot'] + edu['m_dngts_tot']
edu = edu.drop(columns=['f_dngts_tot','m_dngts_tot'])
# rename columns
edu.columns = edu.columns.str.replace('p_', '')
edu.columns = edu.columns.str.replace('_tot', '')
edu
```

```python
edu = edu.apply(pd.to_numeric, errors='coerce')
edu.dtypes
```

```python
edu['tot'] = edu.iloc[:,1:-1].sum(axis=1)
edu
```

```python
# average for people who complete year 8 or below is set to be 7.
edu['avg_year_completeld'] = (edu['y10e'] * 10 + edu['y11e'] * 11 + edu['y12e'] * 12 + edu['y9e'] * 9 + edu['y8b'] * 7) / edu['tot']
edu = edu[['lga_code_2016','avg_year_completeld']]
edu
```

```python
# save to processed dataset
edu.to_csv('../data/curated/sudoData/edu.csv', index=False)
```

```python
age_cols = [" lga_code_2016",' f_tot_15_19', ' f_tot_20_24', ' f_tot_25_34', ' f_tot_35_44', ' f_tot_45_54',
            ' f_tot_55_64', ' f_tot_65_74', ' f_tot_75_84', ' f_tot_85ov',' f_tot_tot','m_tot_15_19',' m_tot_20_24',
            ' m_tot_25_34', ' m_tot_35_44', ' m_tot_45_54', ' m_tot_55_64', ' m_tot_65_74', ' m_tot_75_84', ' m_tot_85ov', ' m_tot_tot']
age = eduLevel[age_cols]
# remove the space in the column names 
age.columns = age.columns.str.replace(' ', '')
age
```

```python
from statistics import mean 
# age between 15 ~ 19 is ignored because a citizen has to be over 18 to be able to vote
# average for 85 over is set to be 80.
# the average age calculated here is only the average age for voters.
age['avg_age_voters'] = ((age['f_tot_20_24']+age['m_tot_20_24']) * mean([10, 24]) + (age['f_tot_25_34']+age['m_tot_25_34']) * mean([35, 34]) 
                  + (age['f_tot_35_44']+age['m_tot_35_44']) * mean([35, 44]) + (age['f_tot_45_54']+age['m_tot_45_54']) * mean([45, 54]) 
                  + (age['f_tot_55_64']+age['m_tot_55_64']) * mean([55, 64]) + (age['f_tot_65_74']+age['m_tot_65_74']) * mean([65, 74]) 
                  + (age['f_tot_75_84']+age['m_tot_75_84']) * mean([75, 84]) 
                  + (age['f_tot_85ov']+age['m_tot_85ov']) * 80) / (age['m_tot_tot'] + age['f_tot_tot'] - age['f_tot_15_19'] - age['m_tot_15_19'])
age = age[['lga_code_2016', 'avg_age_voters']]
age
```

```python
# save to processed dataset
age.to_csv('../data/curated/sudoData/age.csv', index=False)
```
