# R_7 Clean Politics Data

```python
import pandas as pd
```

```python
raw_data_2022 = pd.read_csv('../data/raw/19 and 22 2 candidates/2022_2_candidates.csv')
raw_data_2019 = pd.read_csv('../data/raw/19 and 22 2 candidates/2019_2_candidates.csv')
```

```python
raw_data_2022
```

```python
# Select just the columns wee need to keep, from the raw 2022 election results
df_list = []
for id, data in raw_data_2022.groupby(['DivisionId']):
    
    data.index = [0, 1]

    state = data.loc[0, 'State']
    DivisionId = data.loc[0, 'DivisionId']
    DivisionAb = data.loc[0, 'DivisionAb']
    DivisionName = data.loc[0, 'DivisionName']

    party1 = data.loc[0, 'PartyAb']
    party2 = data.loc[1, 'PartyAb']
    party1_votes = data.loc[0, 'PreferenceVotes']
    party2_votes = data.loc[1, 'PreferenceVotes']

    winning_percent = max(party1_votes, party2_votes)/(party1_votes+party2_votes)

    winning_party = party2
    if party1_votes > party2_votes:
        winning_party = party1

    df_list.append(pd.DataFrame({'state': [state], 'DivisionId': [DivisionId], 'DivisionAb': [DivisionAb], 'DivisionName': [DivisionName],
                                 'WinningParty2022': [winning_party], 'WinningPercent2022': [winning_percent]}))
    
cleaned_data = pd.concat(df_list)
cleaned_data
```

```python
# Select just the columns wee need to keep, from the raw 2019 election results
df_list = []
for id, data in raw_data_2019.groupby(['DivisionId']):
    
    data.index = [0, 1]

    state = data.loc[0, 'State']
    DivisionId = data.loc[0, 'DivisionId']
    DivisionAb = data.loc[0, 'DivisionAb']
    DivisionName = data.loc[0, 'DivisionName']

    party1 = data.loc[0, 'PartyAb']
    party2 = data.loc[1, 'PartyAb']
    party1_votes = data.loc[0, 'PreferenceVotes']
    party2_votes = data.loc[1, 'PreferenceVotes']

    winning_percent = max(party1_votes, party2_votes)/(party1_votes+party2_votes)

    winning_party = party2
    if party1_votes > party2_votes:
        winning_party = party1

    df_list.append(pd.DataFrame({'DivisionAb': [DivisionAb],
                                 'WinningParty2019': [winning_party], 'WinningPercent2019': [winning_percent]}))
    
cleaned_data2 = pd.concat(df_list)
cleaned_data2
```

```python
# merge these two
merged = cleaned_data.merge(cleaned_data2, how='outer', on='DivisionAb')
```

```python
# STIRLING electorate removedin 2022 election
merged = merged[merged['DivisionAb'] != 'STIR']
merged
```

```python
# HAWKE electorate added in 2022
merged.loc[149, 'WinningParty2019'] = '-'
merged.loc[149, 'WinningPercent2019'] = 0.5
merged
```

```python
# create extra columns
merged['ChangedWinningParty'] = merged['WinningParty2022'] != merged['WinningParty2019']
merged['ChangeInWinningPercent'] = merged['WinningPercent2022'] - merged['WinningPercent2019']
merged
```

```python
# add 2022 winning party's 2019 vote percent
df_list = []
for id, data in raw_data_2019.groupby(['DivisionId']):
    
    data.index = [0, 1]

    data['2022WinningParty2019percent'] = data['PreferenceVotes']/sum(data['PreferenceVotes'])

    df_list.append(data[['DivisionAb', 'PartyAb', '2022WinningParty2019percent']])
    
data_2019 = pd.concat(df_list)
```

```python
# again, make changes for Hawke
merged2 = merged.merge(data_2019, how='left', left_on = ['DivisionAb', 'WinningParty2022'], right_on = ['DivisionAb', 'PartyAb'])
merged2.loc[149, 'PartyAb'] = '-'
merged2.loc[149, '2022WinningParty2019percent'] = 0.5
merged2
```

```python
# fill any potential nas with 0.5 (e.g. 2019 party wasn't even 2nd)
merged2['2022WinningParty2019percent'] = merged2['2022WinningParty2019percent'].fillna(0.5)
# create the desired column: how much more votes did winning party in 2022 get compared to 2019
merged2['2022WinningPartyOwnChangeInPercent'] = merged2['WinningPercent2022'] - merged2['2022WinningParty2019percent']
merged2 = merged2.drop(['PartyAb', '2022WinningParty2019percent'], axis = 1)
merged2
```

```python
merged2.to_csv('../data/curated/ElectionData/2022ElectionData.csv')
```
